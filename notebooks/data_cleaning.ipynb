```python
# we are filling median of income in null values
df['Income'].fillna(df['Income'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# converting Dt_Customer to datetime format
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
```

**Feature Engineering**

```python
# creating a new column 'Age' from 'Year_Birth'
df['Age'] = 2025 - df['Year_Birth']
```

```python
# creating a new column 'Family_Size'
df['Family_Size'] = 1 + df['Kidhome'] + df['Teenhome']
```

```python
# creating a new column 'Total_Spending'
spend_cols = ['MntWines','MntFruits','MntMeatProducts','MntFishProducts',
              'MntSweetProducts','MntGoldProds']
df['Total_Spending'] = df[spend_cols].sum(axis=1)
```

```python
# creating a new column 'Total_Purchases'
purchase_cols = ['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']
df['Total_Purchases'] = df[purchase_cols].sum(axis=1)
```

```python
# creating a new column 'Customer_For_Days'
df['Customer_For_Days'] = (df['Dt_Customer'].max() - df['Dt_Customer']).dt.days
```

**Removing illagical data**

```python
# removing rows where Total_Spending is 0
df = df[df['Total_Spending'] > 0]
```

```python
# removing unrealistic ages
df = df[(df['Age'] >= 18) & (df['Age'] <= 90)]
```

```python
# removing income outliers (top 1%)
df = df[df['Income'] < df['Income'].quantile(0.99)]
```

```python
# standardizing Marital_Status values
df['Marital_Status'].replace({
    'YOLO': 'Single',
    'Absurd': 'Single',
    'Alone': 'Single',
}, inplace=True)
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({
    'Together': 'Married'
})
```

```python
# standardizing Education values
df['Education'] = df['Education'].replace({
    '2n Cycle': 'Secondary'
})
```

```python
# Columns to drop
drop_cols = [
    'Z_CostContact',
    'Z_Revenue',
    'Year_Birth',       
    'Dt_Customer'       
]

 
df = df.drop(columns=[col for col in drop_cols if col in df.columns])

df.head()
```

**Saving the Dataset**

```python
import os

save_path = "data/processed"
os.makedirs(save_path, exist_ok=True)

file_path = os.path.join(save_path, "customer_cleaned.csv")
df.to_csv(file_path, index=False)

print("File saved at:", file_path)
```
